# ì‹¤ìŠµë¬¸ì œ 1


 XOR ë¬¸ì œëŠ” ì…ë ¥ê°’ ë‘ ê°œ (0 ë˜ëŠ” 1)ë¥¼ ë°›ì•„ì„œ XOR ì—°ì‚°ì„ ìˆ˜í–‰í•œ ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ëŠ” ë¬¸ì œë¡œ, ì´ë¥¼ ì„ í˜• ë¶„ë¥˜ë¡œëŠ” í•´ê²°í•  ìˆ˜ ì—†ëŠ” ë¬¸ì œë¡œ ì•Œë ¤ì ¸ ìˆìŠµë‹ˆë‹¤. ì´ ë¬¸ì œë¥¼ ë¹„ì„ í˜• í™œì„±í™” í•¨ìˆ˜ì¸ ì‹œê·¸ëª¨ì´ë“œë¥¼ ì‚¬ìš©í•œ ë‹¤ì¸µ í¼ì…‰íŠ¸ë¡ ìœ¼ë¡œ í•´ê²°í•˜ë ¤ê³  í•©ë‹ˆë‹¤. ì•Œë§ê²Œ ì½”ë“œë¥¼ ì±„ì›Œ ë„£ìœ¼ì„¸ìš”.

```python
import torch
import torch.nn as nn

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# for reproducibility
torch.manual_seed(777)
if device == 'cuda':
    torch.cuda.manual_seed_all(777)

X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)

# nn layers

model = nn.Sequential(
          nn.Linear(2, 10, bias=True), # input_layer = 2, hidden_layer1 = 10
          nn.Sigmoid(),
          nn.Linear(10, 1, bias=True), # hidden_layer3 = 10, output_layer = 1
          nn.Sigmoid()
          ).to(device)

# define cost/loss & optimizer
criterion = torch.nn.BCELoss().to(device) # binary cross entrophy
optimizer = torch.optim.SGD(model.parameters(), lr=1)  # modified learning rate from 0.1 to 1

for epoch in range(10001):
    optimizer.zero_grad()

    # forward ì—°ì‚°
    hypothesis = model(X)

    # ë¹„ìš© í•¨ìˆ˜
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()

    if epoch % 1000 == 0:
        print(epoch, cost.item())

```

# ì‹¤ìŠµë¬¸ì œ 2


ì•„ë˜ì˜ ì½”ë“œëŠ” MLP ëª¨ë¸ì„ êµ¬í˜„í•˜ê³ , Test ë° Validationì„ ìˆ˜í–‰í•©ë‹ˆë‹¤. í•˜ì§€ë§Œ ì´ ì½”ë“œì—ëŠ” ëª‡ ê°€ì§€ ë¶€ì¡±í•œ ë¶€ë¶„ì´ ìˆìŠµë‹ˆë‹¤.

ë¹¨ê°„ìƒ‰ìœ¼ë¡œ í•˜ì´ë¼ì´íŠ¸ ëœ ì½”ë©˜íŠ¸ë¥¼ ì°¸ê³ í•˜ì—¬ ì½”ë“œë¥¼ ì™„ì„±ì‹œí‚¤ê³ ,  ì´ ì™¸ì—ë„ ì½”ë“œì— ì˜ëª» ì‘ì„±ëœ ë¶€ë¶„ì´ ìˆë‹¤ë©´ ê³ ì³ì£¼ì„¸ìš”.

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# for reproducibility
random.seed(777)
torch.manual_seed(777)
if device == 'cuda':
    torch.cuda.manual_seed_all(777)

# parameters
learning_rate = 0.001
training_epochs = 15
batch_size = 100
drop_prob = 0.4

# MNIST dataset
mnist_train = dsets.MNIST(root='MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)

mnist_test = dsets.MNIST(root='MNIST_data/',
                         train=False,
                         transform=transforms.ToTensor(),
                         download=True)

# dataset loader
data_loader = torch.utils.data.DataLoader(dataset=mnist_train,
                                          batch_size=batch_size,
                                          shuffle=True,
                                          drop_last=True)

# nn layers
linear1 = torch.nn.Linear(784, 512, bias=True)
linear2 = torch.nn.Linear(512, 512, bias=True)
linear3 = torch.nn.Linear(512, 512, bias=True)
linear4 = torch.nn.Linear(512, 512, bias=True)
linear5 = torch.nn.Linear(512, 10, bias=True)

#í™œì„±í™”í•¨ìˆ˜ë¥¼ ë°”ê¿”ì„œ Gradient Vanishing/Exploding ë¬¸ì œë¥¼ í•´ê²°í•´ë³´ì„¸ìš”!
relu = torch.nn.ReLU()
dropout = torch.nn.Dropout(p=drop_prob)

#Xavier Initialization ì½”ë“œë¥¼ ì‘ì„±í•˜ì—¬ Gradient Vanishing/Exploding ë¬¸ì œë¥¼ í•´ê²°í•´ë³´ì„¸ìš”!
torch.nn.init.xavier_uniform_(linear1.weight)
torch.nn.init.xavier_uniform_(linear2.weight)
torch.nn.init.xavier_uniform_(linear3.weight)
torch.nn.init.xavier_uniform_(linear4.weight)
torch.nn.init.xavier_uniform_(linear5.weight)

# model
model = torch.nn.Sequential(linear1, relu, dropout,
                            linear2, relu, dropout,
                            linear3, relu, dropout,
                            linear4, relu, dropout,
                            linear5).to(device)


#ë” ì¢‹ì€ ì„±ëŠ¥ì„ ë‚´ëŠ” optimizerë¥¼ ì‚¬ìš©í•´ë³´ì„¸ìš”!
criterion = torch.nn.CrossEntropyLoss().to(device)    # Softmax is internally computed.
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

total_batch = len(data_loader)
model.eval()
for epoch in range(training_epochs):
    avg_cost = 0

    for X, Y in data_loader:
        # reshape input image into [batch_size by 784]
        # label is not one-hot encoded
        X = X.view(-1, 28 * 28).to(device)
        Y = Y.to(device)

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis, Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost / total_batch

    print('Epoch:', '%04d' % (epoch + 1), 'cost =', '{:.9f}'.format(avg_cost))

print('Learning finished')

# Test model and check accuracy
with torch.no_grad():
    model.train()

    # Test the model using test sets
    X_test = mnist_test.test_data.view(-1, 28 * 28).float().to(device)
    Y_test = mnist_test.test_labels.to(device)

    prediction = model(X_test)
    correct_prediction = torch.argmax(prediction, 1) == Y_test
    accuracy = correct_prediction.float().mean()
    print('Accuracy:', accuracy.item())
```
