## Broadcasting

- 자동으로 실행되기 때문에 주의해줘야 함.
- 생각했던 계산이 아님에도 오류없이 진행되기 때문에 나중에 문제 발생 시 어디서 잘못됐는지 찾기 어려움

````python
# Vector + Scalar
m1 = torch.FloatTensor([[1, 2]])
m2 = torch.FloatTensor([3]) # 3 -> [[3, 3]]
print(m1 + m2)  # [[1, 2]] + [[3, 3]]
````

````python
# 2 x 1 Vector + 1 x 2 Vector
m1 = torch.FloatTensor([[1, 2]])
m2 = torch.FloatTensor([[3], [4]])
print(m1 + m2)  # [[1, 2], [1, 2]] + [[3, 3], [4, 4]]
````

## Multiplication vs Matrix Multiplication

````python
# matmul
m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print("Shaoe of Matrix m1 :", m1.shape) # 2 x 2
print("Shaoe of Matrix m2 :", m2.shape) # 2 x 1
print(m1.matmul(m2)) # => 2 x 1
````

````python
# mul
m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print("Shaoe of Matrix m1 :", m1.shape) # 2 x 2
print("Shaoe of Matrix m2 :", m2.shape) # 2 x 1
print(m1 * m2)
print(m1.mul(m2)) # => 2 x 2
````

# Lab02 : Linear regresssion

````python
# 학습 데이터 선언
X_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
````

````python
# 초기의 가중치와 편향 설정 -> 주로 0으로 두고 학습 시작
W = torch.zeros(1, requires_grad=True) # requires_grad=True : 학습할 변수임을 명시
b = torch.zeros(1, requires_grad=True)
````

````python
# 옵티마이저 정의 -> SGD 선언
optimizer = torch.optim.SGD([W, b], lr=0.01)
````

````python
# 학습 진행
nb_epochs = 1000
for epoch in range(1, nb_epochs+1):
    # 가설 설정
    hypothesis = W * X_train + b
    # 손실함수 설정 -> 선형 회귀는 주로 MSE 많이 사용
    cost = torch.mean((hypothesis - y_train) ** 2)
    # SGD 선언
    optimizer.zero_grad() # gradient를 0으로 초기화
    cost.backward()       # gradient 계산
    optimizer.step()      # Gradient Descent를 진행하여 W와 b를 개선
````

````python
print(hypothesis) # 예측값 저장되어 있음
print(cost) # 오차 0.0006
````

# Lab03 : Gradient Descent

## 1) SGD 수식으로 풀이

````python
# 학습 데이터
X_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[1], [2], [3]])

# 가중치 초기화
W = torch.zeros(1)

# learning rate 설정
lr = 0.1

# 학습 진행
nb_epochs = 10
for epoch in range(1, nb_epochs+1):
    # 가설 설정,  H(x) 계산
    hypothesis = W * X_train

    # 손실함수와 옵티마이저 설정 -> 선형 회귀는 손실함수로 MSE 많이 사용
    cost = torch.mean((hypothesis - y_train) ** 2)
    gradient = torch.sum((W * X_train - y_train) * X_train)

    print("Epoch {:4d}/{} W : {:.3f}, Cost: {:.6f}".format(
        epoch, nb_epochs, W.item(), cost.item()))

    # cost gradient로 H(x) 개선
    W -= lr * gradient
````

## 2) torch 내장 함수로 풀이

````python
# 학습 데이터
X_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[1], [2], [3]])

# 가중치 초기화
W = torch.zeros(1, requires_grad=True)

# 옵티마이저 정의 -> SGD 선언
optimizer = torch.optim.SGD([W], lr=0.1)

# 학습 진행
nb_epochs = 10
for epoch in range(1, nb_epochs+1):
    # 가설 설정,  H(x) 계산
    hypothesis = W * X_train

    # 손실함수 설정 -> 선형 회귀는 손실함수로 MSE 많이 사용
    cost = torch.mean((hypothesis - y_train) ** 2)

    print("Epoch {:4d}/{} W : {:.3f}, Cost: {:.6f}".format(
        epoch, nb_epochs, W.item(), cost.item()))

    # cost gradient로 H(x) 개선
    optimizer.zero_grad() # gradient를 0으로 초기화
    cost.backward()       # gradient 계산
    optimizer.step()      # Gradient Descent를 진행하여 W와 b를 개선
````

````python
print(hypothesis)
print(cost)
````
