# 이론문제

O,X문제입니다.

```python
1. 클래스가 3인 데이터를 학습시킬 예정이다. |X_train|=(10, 4), |y_train|=(10,) 일 때,  y_train에서 원-핫 인코딩을 진행했다.
   이때 W(가중치)의 크기는 (4, )으로 설정해야 한다.
    -> X, 클래스가 3인 y_train을 원-핫 인코딩을 진행한 결과는 (10, 3)이 되므로 W는 (4, 3)으로 설정해야 한다.
2. Overfitting 발생 시 은닉층을 줄여본다.
    -> O, 은닉층이 깊어질수록 학습의 깊이도 깊어져 과적합이 더 심해질 수 있으므로 줄어보는게 좋다.
3. 1000개의 훈련 데이터가 있을 때, batch size를 500으로 하면 2번의 epoch를 통해 1번의 iteration이 진행된다.
    -> X, 2번의 iteration을 통해 1번의 epoch가 진행된다고 하는게 맞다.
4. 이미지는 (채널, 높이, 넓이)의 형식으로 이루어져 있어서 파이토치 형식에 맞는 (높이, 넓이, 채널)로 변경해줘야 한다.
    -> X, 이미지가 (높이, 넓이, 채널), 파이토치가 (채널, 높이, 넓이)이다.
```

# 실습문제

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(1)

x_train = [[100, 2, 31, 0.1],
           [232, 1, 63, 0.2],
           [341, 1, 13, 0.4],
           [114, 1, 56, 0.5],
           [93, 7, 43, 0.5],
           [301, 2, 24, 0.6]]
y_train = [2, 0, 1, 0, 1, 2]
x_train = torch.FloatTensor(x_train)
y_train = torch.LongTensor(y_train)
```

## 1번

softmax를 이용한 모델 학습 전, 데이터 전처리를 진행할 것입니다. x_train의 분포를 보면 특징들에 따라 값의 분포가 제각각입니다. 또한, y_train을 보면 클래스가 3이라는 것을 알 수 있습니다. 이제 아래의 문제들을 통해 전처리를 진행해 봅시다.

```python
#1-1번 Standardizationd을 진행하세요

mu = '''이부분을 채워보세요'''
sigma = '''이부분을 채워보세요'''

std_x_train = '''이부분을 채워보세요'''
print(std_x_train)
```

```python
# 정답
mu = x_train.mean(dim=0)
sigma = x_train.std(dim=0)

std_x_train = (x_train - mu) / sigma
```

```python
# 1-2번. view를 이용하여 원-핫 인코딩을 진행하세요.

# 레이블에 대한 원-핫 인코딩
y_one_hot = '''이부분을 채워보세요'''
y_one_hot.scatter_('''이부분을 채워보세요''')

print(y_one_hot)
print(y_one_hot.shape())
```

```python
# 정답
y_one_hot = torch.zeros(6, 3) # 모든 원소가 0의 값을 가진 4 × 6 텐서 생성
y_one_hot.scatter_(1, y_train.view(-1, 1), 1) # y텐서의 크기가 (6,) -> (6 x 1)텐서가 됨

print(y_one_hot)
print(y_one_hot.shape)
```

## 2번

무사히 데이터 전처리를 완료하셨군요! 이제 전처리한 std_x_train과 y_one_hot를 갖고서 모델 학습을 진행해 봅시다. 다음 소프트맥스 함수를 이용한 학습 코드를 완성하세요.

```python
# 모델 초기화
W = torch.zeros('''이부분을 채워보세요''')
b = torch.zeros('''이부분을 채워보세요''')

# optimizer 설정
optimizer = optim.SGD([W, b], lr=0.1)

# 학습
nb_epochs = 1000
for epoch in range(nb_epochs + 1):

    # 가설
    hypothesis = F.softmax('''이부분을 채워보세요''')

    # 비용 함수
    cost = '''이부분을 채워보세요'''

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```

```python
# 정답
# 모델 초기화
W = torch.zeros((4, 3), requires_grad=True)
b = torch.zeros((1, 3), requires_grad=True)

# optimizer 설정
optimizer = optim.SGD([W, b], lr=0.1)

# 학습
nb_epochs = 1000
for epoch in range(nb_epochs + 1):

    # 가설
    hypothesis = F.softmax(std_x_train.matmul(W) + b, dim=1)

    # 비용 함수
    cost = (y_one_hot * -torch.log(hypothesis)).sum(dim=1).mean()

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, nb_epochs, cost.item()
        ))
```
